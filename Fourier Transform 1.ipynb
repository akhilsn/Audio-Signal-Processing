```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load audio file in player
audio = "6761.wav"
ipd.Audio(audio)
```

```python
# load audio file in numpy format
signal, sr = librosa.load(audio)
```

```python
# plot waveform
plt.figure(figsize=(20,6))
librosa.display.waveplot(signal, sr=sr, alpha = 0.5)
plt.show()
```

```python
# derive spectrum using FT (Fast Fourier Transform)
ft = sp.fft.fft(signal)
magnitude = np.absolute(ft)
frequency = np.linspace(0, sr, len(magnitude))
```

```python
# plot spectrum
plt.figure(figsize=(30,10))
plt.plot(frequency[:5000], magnitude[:5000])    #magnitude spectrum
plt.xlabel("Frequecy (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

```python
# zoom into the waveform
samples = range(len(signal))
t = librosa.samples_to_time(samples, sr=sr)

plt.figure(figsize=(16,6))

#considering 400 samples
plt.plot(t[10000:10400], signal[10000:10400])
plt.xlabel("Time in sec")
plt.ylabel("Amplitude")
plt.show()
```

```python
# create a sinusoid
# sine Wave -> sin(2*pi(ft - phi))

f = 86
phi = 0
sin = 0.1*np.sin(2*np.pi * (f*t - phi))

plt.figure(figsize=(16,6))

#considering 400 samples

plt.plot(t[10000:10400], sin[10000:10400], color = "r")
plt.xlabel("Time in sec")
plt.ylabel("Amplitude")
plt.show()
```

```python
# compare signal and sinusoid

f = 86
phi = 0
sin = 0.1*np.sin(2*np.pi * (f*t - phi))

plt.figure(figsize=(16,6))

#considering 400 samples
plt.plot(t[10000:10400], signal[10000:10400])
plt.plot(t[10000:10400], sin[10000:10400], color = "r")
plt.fill_between(t[10000:10400], sin[10000:10400]*signal[10000:10400], color = "y")
plt.xlabel("Time in sec")
plt.ylabel("Amplitude")
plt.show()
```
